```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def my_function(x):
    return np.sin(x) / (np.cos(x) ** 2 + np.sqrt(x))
x = np.linspace(-10, 10, 100)
y = my_function(x)
plt.figure(figsize=(10, 5))
plt.plot(x, y, linestyle='--', color='green', alpha=0.3)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('График функции')
plt.grid(True, linestyle='dotted')
plt.text(1, 0.8, 'Вот такая моя функция')
plt.show()
```

```python
X = np.random.normal(0, 1, 3000)
Y = np.random.normal(3, 4, 3000)
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, alpha=0.2, marker='^', s=20, color='purple')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Распределение точек')
plt.grid(True, linestyle='dotted')
plt.show()
```

```python

data = np.random.normal(16, 2, 1000)
plt.hist(data, bins=50, color='red', alpha=0.5)
plt.xlabel('Время забега')
plt.ylabel('N-ов участников')
plt.title('Гистограмма результатов забега школьников')
plt.show()
```
