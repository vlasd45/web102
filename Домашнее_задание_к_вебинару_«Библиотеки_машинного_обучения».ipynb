```python


from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # один признак
y = 4 + 3 * X + np.random.randn(100, 1)  # целевая переменная с небольшим шумом
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)

```

```python
# Создание и обучение модели линейной регрессии
model = LinearRegression()
model.fit(X_train, y_train)
# Предсказание для тестового набора
y_pred = model.predict(X_test)
```

```python
plt. figure(figsize=(12, 6))
plt.scatter(X_train, y_train, color='red', label="Набор обучения")
plt.scatter(X_test, y_test, color='black', label='Тест', alpha=0.9)
# График линейной регрессии
plt.plot(X_test, y_pred, color='green', label="Линейная регрессия")
plt.xlabel('Признак')
plt.ylabel("Целевая переменная")
plt.title("Линейная регрессия: Обучающий и тестовый наборы")
plt.legend()
plt.grid(True)
plt.show()
```
