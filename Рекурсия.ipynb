```python

def recursive_max(numbers, maximum=float('-inf')):
    if not numbers:  # Если нет чисел
        return maximum
    else:
        first_number = numbers[0]  # 1 число из списка
        if first_number > maximum:
            maximum = first_number  # Придаём значение числа моему максимуму
        return recursive_max(numbers[1:], maximum) # вызываю функцию


```

```python

input_list = [3, 4, 5, 6, 7]
print(recursive_max(input_list))
```
